```python
titanic["Age"].fillna(titanic["Age"].median(), inplace= True)
titanic.drop("Cabin",axis=1, inplace= True)
titanic.dropna(subset="Embarked", inplace= True)
```

# **VISUALIZING AND REMOVING OUTLIERS USING BOXPLOT**

```python
sns.boxplot(titanic['Embarked'],color='violet')
```

```python

def removal_box_plot(titanic, column, threshold):
    sns.boxplot(titanic[column],color='Green')
    plt.title(f'Original Box Plot of {column}')
    plt.show()
    removed_outliers = titanic[titanic[column] <= threshold]
    sns.boxplot(removed_outliers[column],color='Green')
    plt.title(f'Box Plot without Outliers of {column}')
    plt.show()
    return removed_outliers
threshold_value = 65
no_outliers = removal_box_plot(titanic, 'Age', threshold_value)
```
