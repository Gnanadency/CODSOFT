# **TITANIC DATASET **

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
titanic=pd.read_csv('/content/archive (4) (1).zip')
```

```python
titanic.info()
```

```python
titanic.shape
```

```python
print(titanic.columns)
```

```python
titanic.describe()
```

```python
titanic.isna().sum()
```

```python
titanic.head(10)
```

```python
titanic.tail(10)
```

```python
titanic.drop(columns='Cabin',inplace=True)
```

```python
titanic.select_dtypes('object').nunique()
```

```python
titanic.drop(columns= ['Name','Ticket'],inplace=True)
```

```python
titanic['Fare'].value_counts()
```

```python
titanic.info()
```

```python
titanic['Age_group']=pd.cut(titanic['Age'],bins=[0,18,35,50,100],labels=['0-18','19-35','36-50','51+'])
```

```python
titanic['Family_Size']=titanic['SibSp']+titanic['Parch']+1
```

```python
titanic.Family_Size.value_counts()
```

```python
titanic['Fare_range']=pd.qcut(titanic['Fare'],q=4,labels=['Low','Medium','High','Veryhigh'])
```

```python
titanic['Fare_range']
```

```python
titanic.head()
```

# **EXPLORATORY DATA ANALYSIS**(EDA)

```python
corr = titanic.select_dtypes("number").drop(columns='Survived').corr()
corr
sns.heatmap(corr, annot=True);
```

```python
Survived_counts=titanic['Survived'].value_counts()
Survived_counts
```

```python
plt.figure(figsize=(20, 6))
explode = (0,0.05)
Survived_counts.plot(kind = 'pie', fontsize = 12, explode = explode, autopct = '%.1f%%')
plt.axis('equal')
plt.title('Survival Representation')
plt.xlabel('Survived')
plt.ylabel('Counts')

plt.show()
```

```python
sex_counts=titanic['Sex'].value_counts()
sex_counts
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x=sex_counts.index,y=sex_counts.values,palette='viridis')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Sex_Count');
```

```python
gender=sns.FacetGrid(titanic,col='Survived')
gender.map(plt.hist,'Sex',bins=2,color='orange')
```

```python
pclass_counts=titanic['Pclass'].value_counts()
pclass_counts
```

```python
plt.figure(figsize=(20, 6))
pclass_counts.plot(kind = 'pie',fontsize = 14,explode=explode, autopct = '%.1f%%')
plt.axis('equal')
explode = (0,0,0.05)
plt.title('Passenger class Representation',fontsize = 14)
plt.xlabel('Pclass',fontsize = 14, labelpad = 20)
plt.ylabel('Counts',fontsize = 14, labelpad = 20)

plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=titanic, x='Pclass', hue='Survived')
plt.title('Survival Count by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(['Not Survived', 'Survived'])
plt.show()
```

```python
Embarked_counts=titanic['Embarked'].value_counts()
Embarked_counts
```

```python
plt.figure(figsize=(20, 6))
Embarked_counts.plot(kind = 'pie',fontsize = 12,explode=explode,autopct = '%.1f%%')
plt.axis('equal')
explode = (0,0,0.05)
plt.title('Embarked Representation')
plt.xlabel('Embarked')
plt.ylabel('Counts')
plt.legend(['A','B','C'])

plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=titanic, x='Embarked', hue='Survived')
plt.title('Embarked Count by Embarked Class')
plt.xlabel('Embarked Class')
plt.ylabel('Count')
plt.legend(['Not Survived', 'Survived'])
plt.show()
```
